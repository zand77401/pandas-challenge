```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#District Summary code

total_schools = school_data_complete['school_name'].nunique()

total_students = school_data_complete['Student ID'].count()



total_budget = school_data_complete['budget'].unique().sum()

avg_math_score = school_data_complete['math_score'].mean()

avg_read_score = school_data_complete['reading_score'].mean()

#Passing Math (The percentage of students that passed math)
total_passing_math_students_df = school_data_complete[school_data_complete['math_score'] >= 70 ]
total_passing_math_students_int = total_passing_math_students_df['Student ID'].count()

percent_passing_math = (total_passing_math_students_int / total_students) * 100

#Passing Reading (The percentage of students that passed reading)
total_passing_read_students_df = school_data_complete[school_data_complete['reading_score'] >= 70 ]
total_passing_read_students_int = total_passing_read_students_df['Student ID'].count()

percent_passing_read = (total_passing_read_students_int / total_students) * 100

#Overall Passing (The percentage of students that passed math and reading.)
overall_passing_df = total_passing_math_students_df[total_passing_math_students_df['reading_score'] >= 70 ]

overall_passing_int = overall_passing_df['Student ID'].count()

percent_overall_passing = (overall_passing_int / total_students) * 100

#Created a dictionary of the above results
dict_district = {'Total Number of Schools': [total_schools], 
                 'Total Number of Students': [total_students], 
                 'Total Budget': [total_budget], 
                 "Average Math Score": [avg_math_score], 
                 'Average Reading Score': [avg_read_score],
                 'Percentage of >= 70 Math Scores': [percent_passing_math],
                 'Percentage of >= 70 Reading Scores': [percent_passing_read],
                 'Percentage of Overall Passing': [percent_overall_passing]
                }
#Created a dataframe from the dictionary of above results
district_df = pd.DataFrame(dict_district)

#Display the results of new district dataframe
district_df
```
